```python
# Option 1: Query current Data

import requests as r
import json
from requests.structures import CaseInsensitiveDict

headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"
headers["Content-Type"] = "application/json"

url = "http://127.0.0.1:8000/query"


text = "Feel good story"


response = r.post(url,
                  headers = headers,
                  data = json.dumps({"text":text}))

candidate = response.json()
for i in range(10):
    print()
    score = candidate['scores'][i]
    doc_id = candidate['corpus_id'][i]
    print(f'With score {score} and doc id {doc_id}, Top {i+1} candidate is:\n')
    print(candidate['passage'][i])
    print()
```

```python
# Option 2: Index current elements

import requests as r
import json
from requests.structures import CaseInsensitiveDict
from os import listdir
from os.path import isfile, join

headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"
headers["Content-Type"] = "application/json"

url = "http://127.0.0.1:8000/index"


file_name_list = listdir('data/data_test')

texts = []

for i in range(50):
    
    with open('data/data_test/'+file_name_list[i]) as f:
        document = f.read()
        texts.append(document)


response = r.post(url,
                  headers = headers,
                  data = json.dumps({"texts":texts}))

```
